```python
print("Highest allowed",df['MSFL'].mean() + 3*df['MSFL'].std())
print("Lowest allowed",df['MSFL'].mean() - 3*df['MSFL'].std())
```

```python
# Finding the Outliers
```

```python
df[(df['MSFL'] > 6.65) | (df['MSFL'] < -4.59)]
```

```python
Trimming the outliers
```

```python
new_df = df[(df['MSFL'] < 6.65) & (df['MSFL'] > -4.59)]
new_df
```

```python
# Capping the outliers
```

```python
upper_limit = df['MSFL'].mean() + 3*df['MSFL'].std()
lower_limit = df['MSFL'].mean() - 3*df['MSFL'].std()
```

```python
#Apply the capping
```

```python
df['MSFL'] = np.where(
    df['MSFL']>upper_limit,
    upper_limit,
    np.where(
        df['MSFL']<lower_limit,
        lower_limit,
        df['MSFL']
    )
)
```
